##Load the iris dataset and split it into train and test sets

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
X, y = load_iris(return_X_y=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

##visualizing the data using scatterplots or histograms

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

```python
# Use load_iris to load the Iris dataset
iris = load_iris()

X = iris.data

y = iris.target
```

###Distribution of Classes in the Iris Dataset

```python
# Create a DataFrame for easier visualization
iris_df = pd.DataFrame(data=X, columns=iris.feature_names)
iris_df['target'] = y

# Count the number of instances for each class
class_counts = iris_df['target'].value_counts()

# Use Seaborn to create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=class_counts.index, y=class_counts.values, palette='viridis')
plt.title('Distribution of Classes in the Iris Dataset')
plt.xlabel('Class')
plt.ylabel('Number of Instances')
plt.show()
```

#### Use Seaborn to create a pair plot with histograms on the diagonal


```python
sns.pairplot(iris_df, hue='target', palette='viridis', diag_kind='hist')
plt.show()
```
